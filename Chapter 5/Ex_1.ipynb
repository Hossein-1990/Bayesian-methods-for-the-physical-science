```python
import warnings
warnings.simplefilter(action="ignore", category=FutureWarning)
import pymc3 as pm
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
```

Ex1.png

```python
aa = np.random.normal(loc = 7 , scale = 2 , size = (1,10000))
a = aa.ravel()
n = [8,100,10000]
for i in n:
    data = a[:i]
    with pm.Model():
        m = pm.Normal('m',mu = -50,sigma = 10)
        obs = pm.Normal('obs',mu = m, sigma = 2,observed = data)
        trace = pm.sample(2000)
        trace2 =  pm.sample_prior_predictive(samples=2000)
    sns.histplot(trace['m'], stat = 'density' , kde = 1)
    sns.histplot(trace2['m'] , stat = 'density' , kde = 1, color = 'r')
    plt.legend(['posterior','prior'])
    plt.show()
```
